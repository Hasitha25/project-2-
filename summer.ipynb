# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
```

### 1. In how many cities Summer Olympics is held so far?

```python
df = pd.read_csv('summer.csv')
df = df.groupby('City').size().reset_index(name='Counts').sort_values(by = 'Counts', ascending = False)

print("Total cities :", len(df))
df.plot.bar(x = 'City', y = 'Counts', figsize =(15, 5))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Medal'] == 'Gold']

df = df.groupby('Sport').size().reset_index(name='Gold').sort_values(by = 'Gold', ascending = False)
print(df.head())

df.head().plot.bar(x = 'Sport', y = 'Gold')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Sport').size().reset_index(name='Total Medals').sort_values(by = 'Total Medals', ascending = False) 
print(df.head())
df.head().plot.bar(x = 'Sport', y = 'Total Medals')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Athlete').size().reset_index(name='Medals').sort_values(by = 'Medals', ascending = False)
print(df.head())
df.head().plot.bar(x = 'Athlete', y = 'Medals')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Medal'] == 'Gold']

df = df.groupby('Athlete').size().reset_index(name='Gold Medals').sort_values(by = 'Gold Medals', ascending = False)
print(df.head())

df.head().plot.bar(x = 'Athlete', y = 'Gold Medals')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df = pd.read_csv('summer.csv')
df = df[df['Country'] == 'IND']
df = pd.DataFrame(df['Year'], columns = ['Year']).groupby('Year').size().reset_index(name='Medals').sort_values(by = 'Year', ascending = True)

print("First Medal to India in", df['Year'].iloc[0])
df.plot.bar(x = 'Year', y = 'Medals', figsize = (15,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Event').size().reset_index(name='Players').sort_values(by = 'Players', ascending = False)  
print(df.head())
df.head().plot.bar(x = 'Event', y = 'Players')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Gender'] == 'Women']
df = df[df['Medal'] == 'Gold']
df = df.groupby('Sport').size().reset_index(name='Female Gold Medals').sort_values(by = 'Female Gold Medals', ascending = False)
print(df.head())
df.head().plot.bar(x = 'Sport', y = 'Female Gold Medals')
```
